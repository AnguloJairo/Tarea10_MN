```python
import numpy as np

# Definicin de las matrices
matrices = {
    "a": np.array([[4, 2, 6], [3, 0, 7], [-2, -1, -3]]),
    "b": np.array([[1, 2, 0], [2, 1, -1], [3, 1, 1]]),
    "c": np.array([[1, 1, -1, 1], [1, 2, -4, -2], [2, 1, 1, 5], [-1, 0, -2, -4]]),
    "d": np.array([[4, 0, 0, 0], [6, 7, 0, 0], [9, 11, 1, 0], [5, 4, 1, 1]])
}

# Anlisis de cada matriz
for key, matrix in matrices.items():
    print(f"\nMatriz {key}:")
    print(matrix)
    
    # Calcular el determinante
    try:
        det = np.linalg.det(matrix)
        print(f"Determinante: {det}")
        
        # Verificar si es no singular
        if not np.isclose(det, 0):
            print("La matriz ES no singular (invertible)")
            inv_matrix = np.linalg.inv(matrix)
            print("Matriz inversa:")
            print(inv_matrix)
        else:
            print("La matriz NO es no singular (no es invertible)")
    except np.linalg.LinAlgError:
        print("La matriz no es cuadrada o no se puede calcular su determinante")
```

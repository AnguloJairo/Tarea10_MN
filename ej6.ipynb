```python
import numpy as np

# Matriz 6a
A = np.array([[2, -1, 1],
              [3, 3, 9],
              [3, 3, 5]], dtype=float)

n = A.shape[0]
L = np.eye(n)  # Matriz identidad
U = A.copy()

print("=== Factorización LU para Matriz 6a ===")
print("\nMatriz original A:")
print(A)

# Paso 1: Primera columna (k=0)
print("\nPaso 1: k=0 (primera columna)")
for i in range(1, n):
    L[i,0] = U[i,0] / U[0,0]
    U[i,:] = U[i,:] - L[i,0] * U[0,:]
    print(f"\nOperación para fila {i+1}:")
    print(f"L[{i+1},1] = U[{i+1},1]/U[1,1] = {L[i,0]:.4f}")
    print("U actualizada:")
    print(U)

# Paso 2: Segunda columna (k=1)
print("\nPaso 2: k=1 (segunda columna)")
for i in range(2, n):
    L[i,1] = U[i,1] / U[1,1]
    U[i,1:] = U[i,1:] - L[i,1] * U[1,1:]
    print(f"\nOperación para fila {i+1}:")
    print(f"L[{i+1},2] = U[{i+1},2]/U[2,2] = {L[i,1]:.4f}")
    print("U actualizada:")
    print(U)

print("\nResultado final:")
print("\nMatriz L:")
print(L)
print("\nMatriz U:")
print(U)
print("\nVerificación A = LU:")
print(np.allclose(A, L @ U))
```

```python
# Matriz 6b
A = np.array([[1.012, -2.132, 3.104],
              [-2.132, 4.096, -7.013],
              [3.104, -7.013, 0.014]], dtype=float)

n = A.shape[0]
L = np.eye(n)
U = A.copy()

print("\n\n=== Factorización LU para Matriz 6b ===")
print("\nMatriz original A:")
print(A)

# Eliminación para k=0
for i in range(1, n):
    L[i,0] = U[i,0] / U[0,0]
    U[i,:] = U[i,:] - L[i,0] * U[0,:]
    print(f"\nPaso {i} para k=0:")
    print(f"Multiplicador L[{i+1},1]: {L[i,0]:.6f}")
    print("U actualizada:")
    print(U)

# Eliminación para k=1
for i in range(2, n):
    L[i,1] = U[i,1] / U[1,1]
    U[i,1:] = U[i,1:] - L[i,1] * U[1,1:]
    print(f"\nPaso {i} para k=1:")
    print(f"Multiplicador L[{i+1},2]: {L[i,1]:.6f}")
    print("U actualizada:")
    print(U)

print("\nResultado final:")
print("\nMatriz L:")
print(L)
print("\nMatriz U:")
print(U)
```

```python
# Matriz 6d
A = np.array([[2.1756, 4.0231, -2.1732, 5.1967],
              [-4.0231, 6.0000, 0, 1.1973],
              [-1.0000, -5.2107, 1.1111, 0],
              [6.0235, 7.0000, 0, -4.1561]], dtype=float)

n = A.shape[0]
L = np.eye(n)
U = A.copy()

print("\n\n=== Factorización LU para Matriz 6d ===")
print("\nMatriz original A:")
print(A)

# Eliminación columna por columna
for k in range(n-1):
    print(f"\n--- Columna {k+1} ---")
    for i in range(k+1, n):
        L[i,k] = U[i,k] / U[k,k]
        U[i,k:] = U[i,k:] - L[i,k] * U[k,k:]
        print(f"\nOperación fila {i+1}:")
        print(f"Multiplicador L[{i+1},{k+1}]: {L[i,k]:.6f}")
        print("U actualizada:")
        print(U)

print("\nResultado final:")
print("\nMatriz L:")
print(L)
print("\nMatriz U:")
print(U)
```

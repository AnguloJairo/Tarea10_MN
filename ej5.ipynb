```python
import numpy as np

# Definición de las matrices y vector para el sistema 5a
L1 = np.array([[1, 0, 0], [2, 1, 0], [-1, 0, 1]])
U1 = np.array([[2, 3, -1], [0, -2, 1], [0, 0, 3]])
b1 = np.array([2, -1, 1])

print("=== SISTEMA 5a ===")
print("\nMatriz L (triangular inferior):")
print(L1)
print("\nMatriz U (triangular superior):")
print(U1)
print("\nVector b:")
print(b1)

# Resolución del sistema L(Ux) = b
# Primero resolvemos Ly = b para y
print("\nPaso 1: Resolver Ly = b")
y1 = np.linalg.solve(L1, b1)
print("Solución intermedia y:")
print(y1)

# Luego resolvemos Ux = y para x
print("\nPaso 2: Resolver Ux = y")
x1 = np.linalg.solve(U1, y1)
print("Solución final x:")
print(x1)

# Verificación
print("\nVerificación (L*U*x debería ser igual a b):")
print(np.allclose(L1 @ U1 @ x1, b1))
```

```python
# Definición de las matrices y vector para el sistema 5b
L2 = np.array([[2, 0, 0], [-1, 1, 0], [3, 2, -1]])
U2 = np.array([[1, 1, 1], [0, 1, 2], [0, 0, 1]])
b2 = np.array([-1, 3, 0])

print("\n\n=== SISTEMA 5b ===")
print("\nMatriz L (triangular inferior):")
print(L2)
print("\nMatriz U (triangular superior):")
print(U2)
print("\nVector b:")
print(b2)

# Resolución del sistema L(Ux) = b
# Primero resolvemos Ly = b para y
print("\nPaso 1: Resolver Ly = b")
y2 = np.linalg.solve(L2, b2)
print("Solución intermedia y:")
print(y2)

# Luego resolvemos Ux = y para x
print("\nPaso 2: Resolver Ux = y")
x2 = np.linalg.solve(U2, y2)
print("Solución final x:")
print(x2)

# Verificación
print("\nVerificación (L*U*x debería ser igual a b):")
print(np.allclose(L2 @ U2 @ x2, b2))
```

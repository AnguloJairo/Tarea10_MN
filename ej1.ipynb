```python
import numpy as np

# Definición de las matrices
A1 = np.array([[2, -3], [3, -1]])
B1 = np.array([[1, 5], [2, 0]])

A2 = np.array([[2, -3], [3, -1]])
B2 = np.array([[1, 5, -4], [-3, 2, 0]])

A3 = np.array([[2, -3, 1], [4, 3, 0], [5, 2, -4]])
B3 = np.array([[0, 1, -2], [1, 0, -1], [2, 3, -2]])

A4 = np.array([[2, 1, 2], [-2, 3, 0], [2, -1, 3]])
B4 = np.array([[1, -2], [-4, 1], [0, 2]])

# Realización de las multiplicaciones
result1 = np.matmul(A1, B1)
result2 = np.matmul(A2, B2)
result3 = np.matmul(A3, B3)
result4 = np.matmul(A4, B4)

# Resultados
print("Resultado a:")
print(result1)
print("\nResultado b:")
print(result2)
print("\nResultado c:")
print(result3)
print("\nResultado d:")
print(result4)
```
